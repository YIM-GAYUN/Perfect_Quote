- 상위 150개의 카테고리만 추출한다.
- 중복되는 카테고리는 제거한다.

```python
categoty_counts_df = pd.read_csv('category_counts.csv')
categoty_counts_df
```

```python
# get the top 150 categories
top150_categories = categoty_counts_df.head(150).values
top150_categories = top150_categories[:,0]

# lower top150_categories
top150_categories = [x.lower() for x in top150_categories]
```

```python
# for 문을 돌면서 중심 카테고리 단어가 다른 단어의 subset인 경우, 중심 카테고리 단어로 바꾸어주는 작업을 하자.

# 예를 들어 inspirational가 중심 카테고리 단어인 경우,
# inspirational-quotes는 inspirational로 바꾸는 것이다.
# inspirational-quote도 inspirational로 바꾸는 것이다.

# 이미 데이터는 빈도수가 높은 대로 정렬되어 있기 때문에 앞쪽 인덱스가 뒤쪽 인덱스의 subset인 경우일 것이다.
for i in range(len(top150_categories)):
    for j in range(i, len(top150_categories)):
        if top150_categories[i] in top150_categories[j]:
            top150_categories[j] = top150_categories[i]

top150_categories = list(set(top150_categories))
print(len(top150_categories))
top150_categories
```
