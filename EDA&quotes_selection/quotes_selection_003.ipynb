```python
# 상위 150개의 카테고리가 category_candidates에 있다.
# 아래 함수는 각 데이터의 카테고리가 category_candidates에 있는지 확인하는 함수이다.
# 이 함수가 1을 반환하면 상위 150개의 카테고리가 포함되어 있다는 뜻이다.
# 이 함수가 0을 반환하면 상위 150개의 카테고리가 포함되어 있지 않다는 뜻이다.
def is_in_candidates(category):
    # print(category)
    each_category = category.split(',')
    for each in each_category:
        if each.strip() in category_candidates:
            return 1
    return 0
```

```python
# in_candidates 컬럼을 추가하여 각 데이터가 상위 150개의 카테고리에 포함되어 있는지 확인한다.
df['in_candidates'] = df['category'].apply(is_in_candidates)
```

```python
# 상위 150개의 카테고리에 포함되어 있는 데이터만 선택한다.
selected = df[df['in_candidates'] == 1]
```

```python
# 불필요한 컬럼을 제거한다.
selected = selected.drop(['in_candidates'], axis=1)
selected = selected.drop(['author_name_only'], axis=1)
selected = selected.drop(['token length'], axis=1)
```
